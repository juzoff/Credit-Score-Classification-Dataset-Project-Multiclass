```python
# prompt: how to identify outliers in data2

# Calculate the IQR for each numerical feature
for col in num_cols:
  Q1 = data2[col].quantile(0.25)
  Q3 = data2[col].quantile(0.75)
  IQR = Q3 - Q1

  # Define the lower and upper bounds for outliers
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR

  # Identify outliers
  outliers = data2[(data2[col] < lower_bound) | (data2[col] > upper_bound)]

  # Print the number of outliers for each feature
  print(f"Number of outliers in {col}: {len(outliers)}")

  # You can also print or analyze the outlier values themselves
  #print(f"Outliers in {col}: {outliers[col].values}")
```

```python
import pandas as pd
from scipy.stats import anderson

# List of continuous columns (exclude non-numeric columns)
continuous_columns = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                      'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Num_of_Delayed_Payment',
                      'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Outstanding_Debt',
                      'Credit_Utilization_Ratio', 'Total_EMI_per_month', 'Amount_invested_monthly',
                      'Monthly_Balance']

# Apply Anderson-Darling test for each continuous column
for column in continuous_columns:
    # Drop missing values and run Anderson-Darling test
    data = data5[column].dropna()  # Drop missing values

    # Perform Anderson-Darling test for normality
    result = anderson(data)

    # Print the results
    print(f"Results for {column}:")
    print(f"Test Statistic: {result.statistic}")
    print("Critical Values:")
    for i, critical_value in enumerate(result.critical_values):
        print(f"  {result.significance_level[i]}%: {critical_value}")

    # Compare the test statistic with the critical values at 5% significance level
    if result.statistic > result.critical_values[2]:  # 5% significance level (third critical value)
        print(f"  => The null hypothesis is rejected. Data for {column} does not follow a normal distribution.\n")
    else:
        print(f"  => The null hypothesis is not rejected. Data for {column} follows a normal distribution.\n")
```

```python
from scipy.stats import kruskal

# List of numeric columns to be tested
numeric_columns = ['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts',
                   'Num_Credit_Card', 'Interest_Rate', 'Num_of_Loan', 'Num_of_Delayed_Payment',
                   'Changed_Credit_Limit', 'Num_Credit_Inquiries', 'Outstanding_Debt',
                   'Credit_Utilization_Ratio', 'Total_EMI_per_month', 'Amount_invested_monthly',
                   'Monthly_Balance']

# Group data by Credit_Score and perform Kruskal-Wallis test on each numeric column
for column in numeric_columns:
    print(f"Results for {column}:")

    # Group the data by Credit_Score
    group_standard = data5[data5['Credit_Score'] == 'Standard'][column].dropna()
    group_poor = data5[data5['Credit_Score'] == 'Poor'][column].dropna()
    group_good = data5[data5['Credit_Score'] == 'Good'][column].dropna()

    # Perform Kruskal-Wallis test
    stat, p_value = kruskal(group_standard, group_poor, group_good)

    # Output results
    print(f"Test Statistic: {stat}")
    print(f"P-value: {p_value}")

    # Interpret the result
    if p_value < 0.05:
        print(f"  => The differences between the groups are statistically significant.\n")
    else:
        print(f"  => The differences between the groups are not statistically significant.\n")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# List of categorical columns
categorical_columns = ['Occupation', 'Type_of_Loan', 'Credit_Mix', 'Payment_of_Min_Amount', 'Payment_Behaviour']

# Perform Chi-Square test for each categorical variable against Credit_Score
for column in categorical_columns:
    print(f"Results for {column}:")

    # Create a contingency table
    contingency_table = pd.crosstab(data5['Credit_Score'], data5[column])

    # Perform the Chi-Square test
    chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

    # Output results
    print(f"Chi2 Statistic: {chi2_stat}")
    print(f"P-value: {p_value}")

    # Interpret the result
    if p_value < 0.05:
        print(f"  => The association between {column} and Credit_Score is statistically significant.\n")
    else:
        print(f"  => The association between {column} and Credit_Score is not statistically significant.\n")
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
import numpy as np

# Set k to 5 for imputation
k = 5

# List of columns where you want to impute whole numbers
whole_number_columns = ['Age', 'Num_Bank_Accounts', 'Num_Credit_Card',
                        'Num_of_Loan', 'Num_of_Delayed_Payment', 'Num_Credit_Inquiries']

# Select numeric columns to perform KNN imputation
numeric_columns = data5.select_dtypes(include=['float64', 'int64']).columns

# Standardize the data (optional but recommended for KNN)
scaler = StandardScaler()
data5_scaled = data5.copy()
data5_scaled[numeric_columns] = scaler.fit_transform(data5[numeric_columns])

# Initialize KNNImputer with k=5 and using median strategy for imputation
knn_imputer = KNNImputer(n_neighbors=k, metric='nan_euclidean')

# Apply KNNImputer to the scaled numeric columns
imputed_data = knn_imputer.fit_transform(data5_scaled[numeric_columns])

# Reverse the standardization (inverse transform the imputed data back to the original scale)
imputed_data_original_scale = scaler.inverse_transform(imputed_data)

# Convert the imputed data back to a DataFrame
data5_imputed = data5.copy()
data5_imputed[numeric_columns] = imputed_data_original_scale

# Round the specific columns to whole numbers
data5_imputed[whole_number_columns] = data5_imputed[whole_number_columns].round().astype(int)

# Check for missing values after imputation
missing_values_after_imputation = data5_imputed.isnull().sum()
print(missing_values_after_imputation)

# Check the result after imputation
print(data5_imputed[['Age', 'Num_Bank_Accounts', 'Num_Credit_Card',
                     'Num_of_Loan', 'Num_of_Delayed_Payment', 'Num_Credit_Inquiries',
                     'Monthly_Balance']].head())
```
